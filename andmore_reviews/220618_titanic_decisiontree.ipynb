```python
import pandas as pd

titanic = pd.read_csv('datas/titanic_rev.csv')
titanic.head()
```

```python
new_sex = []
for each in titanic['Sex']:
    if each == 'male':
        new_sex.append(0)
    else:
        new_sex.append(1)
```

```python
titanic['Sex_n'] = new_sex
titanic.head()
```

```python
new_embarked = []
for each in titanic['Embarked']:
    if each == 'S':
        new_embarked.append(0)
    elif each == 'C':
        new_embarked.append(2)
    else:
        new_embarked.append(3)
```

```python
titanic['Embarked_n'] = new_embarked
titanic.head()
```

```python
titanic_pre = titanic[['Pclass', 'Sex_n', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked_n']]
```

```python
titanic_pre = titanic_pre.reset_index()
titanic_pre = titanic_pre.drop(['index'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(titanic_pre, titanic[['Survived']], test_size=0.1, random_state=13)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(max_depth=3, random_state=13)
tree_clf.fit(X_train, y_train)
print('Score: {}'.format(tree_clf.score(X_train, y_train)))
```

```python
from sklearn.metrics import accuracy_score

y_pred = tree_clf.predict(X_test)
print("Test Accuracy is ", accuracy_score(y_test, y_pred)*100)
```

```python
dicaprio = [3., 0., 19., 0., 0., 5., 1]
winslet = [1., 1., 17., 1., 2., 100., 1]

def isSurvived(name, person):
    isSurvive = 'not survived' if tree_clf.predict([person])[0] == 0 else 'survived'
    print(name, ' is ', isSurvive, 
          ' --> ', max(tree_clf.predict_proba([person])[0]))
    
isSurvived('Dicaprio', dicaprio)
isSurvived('Winslet', winslet)
```
